```python
import pandas as pd
from numpy import nan as np_nan
```

```python
from rdkit import Chem
from rdkit.Chem.rdchem import AtomValenceException, KekulizeException
```

```python
from molvs import standardize_smiles

def standartize_all(comps):
    if not comps:
        return ""
    return ".".join([standardize_smiles(i) for i in comps.split(".")])

def standardize_reaction(rxn_smi):
    stripped_smi = rxn_smi.strip('"').split(" ")[0]
    _reactants, _reagents, _prods = stripped_smi.split(">")
    
    try:
        standartized_smi = standartize_all(_reactants) + ">" + standartize_all(
                    _reagents) + ">" + standartize_all(_prods)
    except AtomValenceException:
        print("valence error")
        print(rxn_smi)
        return np_nan
    except KekulizeException:
        print("can't kekulize")
        print(rxn_smi)
        return np_nan
    else:
        return standartized_smi
```

### Preparing dataset A

```python
path_a = "data/visual_validation/dataSetA.csv"
```

```python
ds_a = pd.read_csv(path_a)
ds_a = ds_a[["rxn_Class", "rxn_Smiles"]]
ds_a = ds_a.reindex(columns=ds_a.columns[::-1])
```

```python
ds_a
```

```python
ds_a.rxn_Smiles = ds_a.rxn_Smiles.apply(standardize_reaction)
```

```python
ds_a.dropna(inplace=True)
```

```python
ds_a.shape
```

```python
ds_a.to_csv("data/visual_validation/validation_a.csv", sep=";", header=None, index=None)
```

# Moving all agents to the left in dataset B

```python
path = "data/visual_validation/validation_b.csv"
```

```python
with open(path.rstrip(".csv") + "_agents_on_left.csv", 'w') as here, open(path, 'r') as f:
    for line in f.readlines():
        reag, ag, prod = line.split('>')
        here.write(".".join((reag, ag)) + '>>' + prod)
```
